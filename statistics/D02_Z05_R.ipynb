# Rozkład normalny
Gęstość zmienna losowa $X$ o rozkładzie normalnym z parametrami $\mu$ i $\sigma$ jest dana przez 
$$
f(x) = \frac{1}{\sigma \sqrt{2 \pi} } e^{-\frac{(x-m)^2}{2\sigma^2} }.
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

# Definiujemy zmienną losową

```python
(mu, sigma)=(0,1)
norm=stats.norm(mu, sigma)
```

# Narysujmy gętość zmiennej losowej

```python
t = np.arange(-5, 5, 0.05)
fig, ax = plt.subplots(1, 1)
plt.plot(t, norm.pdf(t), '--', ms=12, label='frozen pdf')
ax.set_xlim([-5,5])
ax.set_ylim([-0.1,1])
plt.show()
```

# Narysujmy dystrybuantę zmiennej losowej

```python
fig, ax = plt.subplots(1, 1)
ax.plot(t, norm.cdf(t), '--', ms=8, label='bernoulli cdf')
plt.show()
```

# Wylosujmy próbkę i narysujmy histogram na jednym obrazku

```python
r = norm.rvs(size=1000)
fig, ax = plt.subplots(1, 1)
ax.hist(r, density=True, histtype='stepfilled', alpha=0.5, label='histogram')
ax.legend(loc='best', frameon=False)
ax.plot(t, norm.pdf(t), 'k-', lw=2, label='m=0, sigma=1')
ax.set_xlim([-5,5])
ax.set_ylim([-0.1,1])
ax.legend()
plt.show()
```

```python
r = norm.rvs(size=1000)
plt.hist(r, density=True, histtype='stepfilled', alpha=0.5, label='histogram')
plt.plot(t, norm.pdf(t), 'k-', lw=2, label='m=0, sigma=1')
plt.xlim([-5,5])
plt.ylim([-0.1,0.5])
plt.show()
```

# Narysuj gęstości rozkładu jednostajnego z różnymi parametrami

```python
sigma_values = [.1,.5,.8]
mu_values = [0.5, 0.5, 0.5]
fig, ax = plt.subplots(1, 1)
for mu, sigma in zip(mu_values, sigma_values):
    dist = stats.norm(mu, sigma)
    plt.plot(t, dist.pdf(t), ls="--", c='black', label=r'$\mu=%1.f,\ \sigma=%.1f$' % (mu, sigma))
ax.set_xlim([-5,5])
plt.legend()
plt.show()
```

# Narysuj kilka histogramów różnych próbek z roskładu jednostajnego o tych samych parametrach.

```python
norm=stats.norm(0,1)
fig, ax = plt.subplots(nrows=5, ncols=5, figsize=(10, 10))

for i in np.arange(0,5):
    for j in np.arange(0,5):
        r = norm.rvs(size=100)
        ax[i,j].hist(r, normed=True)
plt.show()
```
